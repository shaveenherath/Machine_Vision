### 1. Least squares error

```python
#least squares error
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
x = np.linspace(0, 10, 10)
y = 2 * x + 1 + np.random.normal(0, 2, 10)

# create the X matrix with a column of ones
n = len(x)
X = np.concatenate((np.ones((n, 1)), x.reshape(n, 1)), axis=1)
# calculate the least squares solution
B = np.linalg.inv(X.T @ X) @ X.T @ y
print(B.shape)

print("m = " , B[1], ",", "b = " , B[0])

# plot the predicted line
plt.plot(x, 2 * x + 1, 'g', label='True line')
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, X @ B, 'r', label='Predicted line')
plt.legend()
plt.show()

```

### 3. HoughLines

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread(r"D:\ENTC\5th sem\machine vision\open cv\Images\sudoku.jpg")
img = cv.cvtColor(img , cv.COLOR_BGR2RGB)
assert img is not None, "file could not be read, check with os.path.exists()"

gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
edges = cv.Canny(gray, 50, 150, apertureSize=3)
lines = cv.HoughLines(edges, 1, np.pi/180, 200)

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
f , ax = plt.subplots(1 , 3 , figsize=(15 , 10))

ax[0].imshow(img , )
ax[0].set_title('original')
ax[1].imshow(edges)
ax[1].set_title('edges')
ax[2].imshow(gray)
ax[2].set_title('gray')

plt.show()


```
